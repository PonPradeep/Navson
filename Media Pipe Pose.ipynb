# Calculate Angles

```python
def calculate_angle(a,b,c):
    a = np.array(a) # First
    b = np.array(b) # Mid
    c = np.array(c) # End
    
    radians = np.arctan2(c[1]-b[1], c[0]-b[0]) - np.arctan2(a[1]-b[1], a[0]-b[0])
    angle = np.abs(radians*180.0/np.pi)
    
    if angle >180.0:
        angle = 360-angle
        
    return angle 
```

```python
def time_diff(t1,t2):
    t1=t1.split(":")
    t2= t2.split(":")
    time1 = int(t1[0])*60*60 + int(t1[1])*60 + int(t1[2])
    time2 = int(t2[0])*60*60 + int(t2[1])*60 + int(t2[2])
    
    diff= time2-time1
    h,m,s=0,0,0
    if diff>3600:
        h=diff//3600
        diff%=3600
    if diff>60:
        m=diff//60
        diff%=60
    s= diff
    
    return("{}:{}:{}".format(h,m,s))
print(time_diff('21:40:27', '22:40:24'))
```
